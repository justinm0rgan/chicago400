### Census Tract Boundaries

- can join with med rent on GEOID
- WGS 84 4326

```python
census_tracts = gpd.read_file('zip:///Users/justinwilliams/projects/chicago400/data/Boundaries - Census Tracts - 2010.zip')
```

```python
census_tracts.info()
```

```python
census_tracts.crs
```

```python
# join census tracts with med rent
census_merge = pd.merge(left=census_tracts,
        right=census_tract_med_rent,
        left_on='geoid10',
        right_on='GEOID')
```

```python
census_merge.info()
```

```python
# drop unecessary col
census_merge.drop(columns=['geometry_y', 'GEOID'], inplace=True)
```

```python
# rename geometry from census tracts
census_merge.rename(columns={'geometry_x':'geometry'}, inplace=True)
```

```python
census_merge.name = 'census_merge'
```

```python
# list of all shpfiles
shp_files = [bike_routes, daycare, daycare_all, schools_all, cities, forestry, parks, 
             public_schools, school_cook_county, open_spaces, census_merge, loc_aff_df]
```

### Custom function to transform CRS

```python
def set_crs(gdf,to_crs=32616):
    """
    Enter GDF and optional arg for CRS
    Defaults to 32616 for WGS 84 UTM 16N
    """
    gdf.to_crs(to_crs,inplace=True)
    return gdf.head()
```

```python
# apply custom formula to shp_file list
for gdf in shp_files:
    set_crs(gdf)
```

```python
for gdf in shp_files:
    print("{} is in {} crs".format(gdf.name, gdf.crs))
```

```python
# df will not take assignment of name attribute and therefore won't work in loop
chicago_boundaries.to_crs(32616,inplace=True)
```
